```python
# -*- coding: utf-8 -*-
'''如果需要自己构造的话'''

import numpy as np
import matplotlib.pyplot as plt
import matplotlib
import operator
from functools import reduce
import pandas as pd

name = 'informer_HI_pw_quadratic_ftM_sl148_ll148_pl0_lr0.0001_bs24_dm512_nh8_el2_dl1_df2048_atfull_fc5_ebtimeF_dtTrue_mxTrue_test_0'
metrics = np.load('./results/{}/metrics.npy'.format(name))
synthetic = np.load('./results/{}/pred.npy'.format(name))
trues = np.load('./results/{}/true.npy'.format(name))

print(type(synthetic),synthetic.shape)
print(type(trues),trues.shape)

#sensor name
sensor_columns =['sensor 2','sensor 3','sensor 4','sensor 7','sensor 8','sensor 9','sensor 11','sensor 12','sensor 13','sensor 14','sensor 15','sensor 17','sensor 20','sensor 21']+['HI']

print('fakedata:', synthetic.shape[0],'fakedata length:',synthetic.shape[1])
synthetic_df = pd.DataFrame(synthetic.reshape(synthetic.shape[0]*synthetic.shape[1],synthetic.shape[2]),columns=sensor_columns)
true_df = pd.DataFrame(trues.reshape(synthetic.shape[0]*synthetic.shape[1],synthetic.shape[2]),columns=sensor_columns)
print('fakedata_df shape:', synthetic_df.shape)

#添加‘unit_id’这一列
synthetic_df['unit_id'] = np.zeros(synthetic.shape[0]*synthetic.shape[1])
true_df['unit_id'] = np.zeros(synthetic.shape[0]*synthetic.shape[1])
for i in range(synthetic.shape[0]):
    unit_id = 100 + 1 + i
    synthetic_df.loc[i*synthetic.shape[1]:(i+1)*synthetic.shape[1]-1,['unit_id']] = unit_id
    true_df.loc[i*synthetic.shape[1]:(i+1)*synthetic.shape[1]-1,['unit_id']] = unit_id

####np.savetxt('D:/Informer_data_augmentation/fakedata.csv',synthetic_df,delimiter=',')
synthetic_df.set_index('unit_id',inplace=True,drop=True)
true_df.set_index('unit_id',inplace=True,drop=True)

synthetic_df.to_csv('./synthetic_data.csv'.format(name),header=True,index=True)
true_df.to_csv('./true_data.csv'.format(name),header=True,index=True)
```
